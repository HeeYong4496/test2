```python
def max3(a, b, c):
  """ max3 test sample"""
  maximum = a
  if b > maximum: maximum = b
  if c > maximum: maximum = c
  return maximum

print(f'max3(3, 2, 1)= {max3(3, 2, 1)}')
print(f'max3(3, 2, 2)= {max3(3, 2, 2)}')
print(f'max3(3, 1, 2)= {max3(3, 1, 2)}')
print(f'max3(3, 2, 3)= {max3(3, 2, 3)}')
print(f'max3(1, 2, 3)= {max3(1, 2, 3)}')
print(f'max3(2, 3, 1)= {max3(2, 3, 1)}')
```

```python
def med3(a, b, c):
  """a, b, c median value"""

  if a >= b:
    if b >= c:
      return b
    elif a <= c:
      return a
    else:
      return c
  elif a > c:
    return a
  elif b > c:
    return c
  else:
    return b

print("median value")    
a = int(input('int a input=>'))
b = int(input('int b input=>'))
c = int(input('int c input=>'))
print(f'median value={med3(a, b, c)}')

```

```python
# iterable unpacking 의 사용예

gen = ( i ** 2 for i in range(5))
a, b, c, d, e = gen

print(a, b, c, d, e)
# tuple 의 변수의 갯수는 숫자의 갯수와 반드시 일치해야 한다

my_dict = {'one':1, 'two':2, 'three':3}
a, b, c = my_dict
print(a, b, c)

a, b, c = my_dict.values()
print(a, b, c)

a, b, c = my_dict.items()
print(a, b, c)

```

Operator *를 통한 Packing 

operator *는 the tuple (or iterable) unpacking operator

```python
*a, = 1, 2
print(a)
print('\n')

a, *b = 1, 2, 3
print(a)
print(b)
print('\n')

*a, b = 1, 2, 3
print(a)
print(b)
print('\n')

a, *b, c = 1, 2, 3
print(a)
print(b)
print(c)
print('\n')

a, *b, c, d = 1, 2, 3
print(a)
print(b)
print(c)
print(d)
print('\n')
```

```python
gen = (2 ** x for x in range(10))
print(gen)

*g, = gen
print(g)

```

```python
n = int(input("몇개 출력=>"))
w = int(input("줄바꿈 몇개마다?"))

for i in range(n):
  print('*', end='')
  if i % w == w - 1:
    print()

if n % w:
  print()    
```
